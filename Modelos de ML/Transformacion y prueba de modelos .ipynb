```python
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
df_test = pd.read_csv("../datasets/hospitalizaciones_test.csv")
df_train = pd.read_csv("../datasets/hospitalizaciones_train.csv")
```

vemos la correlacion. Vemos la necesidad de transformar algunas columnas a valores numéricos.

```python
df_train.corr()
```

Creamos nueva columna con mas o menos de 8 dias de premanencia en el hospital y eliminamos la anterior.

```python
df_train ['Stay'] = 0
df_train.loc[df_train['Stay (in days)'] > 8, 'Stay'] = 1
df_train.drop(['Stay (in days)'],axis=1, inplace= True)
```

Convertimos variables categoricas a datos numericos para que el modelo las pueda tomar.

```python
le = preprocessing.LabelEncoder()
```

Hcemos una funcion para no repetir codigo

```python
def preproces(x):
    le.fit(x)
        
    return le.transform(x)
```

```python
df_train['Department'] =preproces(df_train['Department'])
df_train['doctor_name'] =preproces(df_train['doctor_name'])
df_train['Age'] =preproces(df_train['Age'])
df_train['gender'] =preproces(df_train['gender'])
df_train['Type of Admission'] =preproces(df_train['Type of Admission'])
df_train['Severity of Illness'] =preproces(df_train['Severity of Illness'])
df_train['health_conditions'] =preproces(df_train['health_conditions'])
df_train['Ward_Facility_Code'] =preproces(df_train['Ward_Facility_Code'])
df_train['Insurance'] =preproces(df_train['Insurance'])
```

Importamos librerias relativas a ml.

```python


from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report
```

Preparamos el dataset de entrenamiento y testeo.

```python
x= df_train.drop(['Stay'], axis= 1)
y= df_train['Stay']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=60, stratify=y)
```

Hacemos prueba con knn, establecemos los parametros.

```python
import scipy
from scipy import stats


param_grid = {'n_neighbors':np.arange(1, 20),
              'weights': ['uniform', 'distance'], 
              'leaf_size':[1,3,5,7,10],
              'algorithm':['auto', 'kd_tree']}
```

```python
knn = KNeighborsClassifier()
```

```python

model2 = GridSearchCV(knn, param_grid=param_grid, cv=5)

# Entrenamos
model2.fit(X_train, y_train)
```

```python
print("Mejores hiperparámetros: "+str(model2.best_params_))
print("Mejor Score: "+str(model2.best_score_)+'\n')

scores = pd.DataFrame(model2.cv_results_)
scores
#Mejores hiperparámetros: {'algorithm': 'auto', 'leaf_size': 1, 'n_neighbors': 19, 'weights': 'uniform'}
#Mejor Score: 0.5958048780487806 mejor tree!
```

Vemos la matriz de confusion.

```python
prediction3 = model2.predict(X_test)


cm = confusion_matrix(y_test,prediction3)
print("Matriz de confusión:")
print(cm)
```
